```python
import cv2
import tensorflow as tf
import numpy as np

model = tf.keras.models.load_model("cnn_helmet_detector.h5")

IMG_SIZE = (128, 128)

classes = ["Tanpa Helm", "Dengan Helm"]

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Webcam tidak terdeteksi.")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        break
    img = cv2.resize(frame, IMG_SIZE)
    img_array = np.expand_dims(img, axis=0) 
    img_array = img_array / 255.0 

    pred = model.predict(img_array)[0][0]
    label = classes[1] if pred >= 0.5 else classes[0]
    color = (0, 255, 0) if pred >= 0.5 else (0, 0, 255)

    cv2.putText(frame, f"{label} ({pred:.2f})", (20, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)
    cv2.rectangle(frame, (10, 10), (300, 60), color, 2)

    cv2.imshow("Helmet Detection - Webcam", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
